```python
def import_from_local_file(path,file_type):
        """
            This function is used to import data from local file system.
            The function takes file path and the file type as arguments/inputs and returns a 
            dataframe (Formats: .csv,.tsv, .xlsx, .json) as output.
            Modules:json, json_normalize
    
        """
        if file_type == 'CSV':
            try:
                df=pd.read_csv(path)
                return df
            except :
                print("Oops!File doesn't exist")
        elif file_type == 'TSV':
            try:
                df=pd.read_table(path)
                return df
            except :
                print("Oops!File doesn't exist")
        elif file_type == 'XLSX':
            try:
                df=pd.read_excel(path)
                return df
            except :
                print("Oops!File doesn't exist")
        elif file_type == 'JSON':
            try:
                df=pd.read_json(path,orient='columns')
                return df
            except :
                print("Oops!File doesn't exist")
        else:
            print("Invalid File Format")
```

```python
df = pd.read_csv("../input/trainset/train.csv",low_memory=False,infer_datetime_format=True)
def EDA():
    b = pd.DataFrame()
    b['Missing value'] = df.isnull().sum()
    b['Missing Percentage']= (b['Missing value']/df.shape[0])*100
    b['N unique value'] = df.nunique()
    b['dtype'] = df.dtypes
    b['max'] = df.max()
    b['min'] = df.min()
    b['mean'] = df.mean()
    b['std_dev'] = df.std()
    b.loc[ (b.dtype == 'object'), 'max'] = 'NA'
    b.loc[ (b.dtype == 'object'), 'min'] = 'NA'
    b.loc[ (b.dtype == 'object'), 'mean'] = 'NA'
    b.loc[ (b.dtype == 'object'), 'std_dev'] = 'NA'
    b.loc[ (b.dtype == 'int64'), 'dtype'] = 'Numeric'
    b.loc[(b.dtype =='float64'),'dtype'] ='Numeric'

    b.loc[ (b.dtype == 'bool'), 'max'] = 'NA'
    b.loc[ (b.dtype == 'bool'), 'min'] = 'NA'
    b.loc[ (b.dtype == 'bool'), 'mean'] = 'NA'
    b.loc[ (b.dtype == 'bool'), 'std_dev'] = 'NA'
    b.loc[ (b.dtype == 'category'), 'max'] = 'NA'
    b.loc[ (b.dtype == 'category'), 'min'] = 'NA'
    b.loc[ (b.dtype == 'category'), 'mean'] = 'NA'
    b.loc[ (b.dtype == 'category'), 'std_dev'] = 'NA'
    b.loc[ (b.dtype == 'datetime64'), 'max'] = 'NA'
    b.loc[ (b.dtype == 'datetime64'), 'min'] = 'NA'
    b.loc[ (b.dtype == 'datetime64'), 'mean'] = 'NA'
    b.loc[ (b.dtype == 'datetime64'), 'std_dev'] = 'NA'
    b.loc[ (b.dtype == 'timedelta[ns]'), 'max'] = 'NA'
    b.loc[ (b.dtype == 'timedelta[ns]'), 'min'] = 'NA'
    b.loc[ (b.dtype == 'timedelta[ns]'), 'mean'] = 'NA'
    b.loc[ (b.dtype == 'timedelta[ns]'), 'std_dev'] = 'NA'

            #print(type(self.b))
    b.dtype = b.dtype.astype(str)
    return b,df
```
